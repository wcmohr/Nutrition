###### Looking at the ratio of animal food to plant foods for means of nutrients

```python
foods_per_calorie = pd.DataFrame({food:foods.loc[food,:]/foods.loc[food,'calories'] for food in foods.index})
foods_per_calorie = food_per_calorie

pf_per_cal = foods_per_calorie.query('cholesterol == 0')
af_per_cal = foods_per_calorie.query('cholesterol != 0')

af_pf_ratio = af_per_cal.mean()/pf_per_cal.mean()
af_pf_ratio.sort_values(ascending=False, inplace=True)
af_pf_ratio
```

From here we see clearly that on aggregate plants have fiber and animal foods do not while animal foods have cholester and plants do not.  Aside from infinite ratios, on a per calorie basis plant foods have much more vitamins K,C and A as well as more manganese, chromium, folate, copper, iron while animal foods have more protein, essenital amino acids, vitamins B12 and D, niacin, choline, chloride and phosphorous.  

```python
af_pf_ratio_max = af_per_cal.max()/pf_per_cal.max()
af_pf_ratio_max.sort_values(ascending=False, inplace=True)
af_pf_ratio_max
```
