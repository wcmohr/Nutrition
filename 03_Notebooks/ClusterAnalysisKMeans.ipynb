```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler as ss
```

```python
foods = pd.read_csv('./Data_formatted/nv_df.csv', index_col='Unnamed: 0')
```

```python
scaler = ss()
```

```python
foods_normalized = pd.DataFrame(scaler.fit_transform(foods))

foods_normalized.index = foods.index
foods_normalized.columns = foods.columns

foods_normalized

# Hypothetical clusters:\
# Grains, nuts/seeds, fruits, vegetables, meat/eggs/dairy

kmeans = KMeans(n_clusters=5)

clusters
pd.set_option('display.max_rows', None)

clusters = kmeans.fit_predict(foods_normalized)
pd.DataFrame.from_records([(clusters[num], food) for num,food in 
    enumerate(foods.index)],columns = ['cluster','food']).sort_values(by='cluster')
```

```python
scaler = StandardScaler()
df_normalized = scaler.fit_transform(foods)

# Determine the range of clusters to try
max_clusters = 10
inertia = []  # Sum of squared distances from each data point to its assigned cluster center

for num_clusters in range(1, max_clusters + 1):
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    kmeans.fit(df_normalized)
    inertia.append(kmeans.inertia_)

# Plot the elbow curve
plt.figure(figsize=(8, 6))
plt.plot(range(1, max_clusters + 1), inertia, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Sum of Squared Distances')
plt.title('Elbow Plot for K-Means Clustering')
plt.grid(True)
plt.show()
# In the elbow plot, the x-axis represents the number of clusters, and the y-axis represents the sum of squared distances. The "elbow" point is the point where the curve starts to level off, indicating that adding more clusters does not significantly reduce the sum of squared distances. The optimal number of clusters is often chosen to be at or near this elbow point.

# Please note that the elbow method is not always definitive, and other techniques, like the silhouette score, can be used to validate the chosen number of clusters. Experiment with different numbers of clusters and choose the one that makes the most sense for your data and problem domain.
```
